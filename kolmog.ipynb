```python
import pandas

import matplotlib.pyplot as plt
dat = [ 3.577,3.778,3.877,3.377,3.857,3.975,3.599,3.492,3.781,3.543,3.986,3.855,3.826,3.293,3.858,3.786]
from scipy.stats import kstest

kstest(dat, 'norm')
```

```python
n = 16

plt.ylim(3.093,3.986)
plt.bar(list(range(n)), dat)
plt.title('Experiment')
plt.show()
```
